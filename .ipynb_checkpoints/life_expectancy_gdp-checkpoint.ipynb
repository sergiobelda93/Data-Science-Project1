DATA

```python
#import libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

#Load Data
df = pd.read_csv('all_data.csv')
print(df.head())
print(df.tail())

# GDP - Gross Domestic Product in U.S. dollars
```

```python
#Change some Column names
df.rename(columns = {'Life expectancy at birth (years)' : 'Life_expectancy'}, inplace = True)

print(df.head())
```

```python
# EXPLORE more your dataset

# .columns gives you the column names
print(df.columns)
print('\n')

# .info() display a summary of the table
print(df.info())
print('\n')

# .describe() display the summary statistics of the table
print(df.describe())
print('\n')

# column_name.value_counts() display the distinct values for a colum
print(df.Country.value_counts())
print(df.Year.value_counts())
print('\n')

## the .shape method in pandas identifies the number of rows and columns in our dataset as (rows, columns)
df.shape
```

```python
# Data wrangling and tidying

## counts the number of missing values in each column 
print(df.isna().sum())

# There's no missing values in the dataframe. In case it was necessary the .drop_duplicates() function removes duplicate rows
# df = df.drop_duplicates() 

```

ANALYSIS. EXPLAIN your data

```python
# 1.0 Which country has the highest life expectancy
sns.set_style('darkgrid')
sns.set_context("notebook", font_scale = 1.2, rc={"grid.linewidth": 2.0, 'axes.titlesize': 30})
plt.figure(figsize=(14, 8))
ax1 = plt.subplot()
sns.violinplot(data=df, x="Country", y="Life_expectancy")
ax1.set_title('Life expectancy per country')
plt.show()

```

```python
# 2.0 Investigate the relationship between Life_expectancy and GDP
# A scatter plot is a great option for investigating the relationship between two quantitative variables. 
plt.figure(figsize=(12, 6))
ax1 = plt.subplot(1, 2, 1)
sns.scatterplot(data = df, x = df.Life_expectancy, y = df.GDP)

ax2 = plt.subplot(1, 2, 2)
sns.scatterplot(data = df, x = df.Life_expectancy, y = df.GDP, hue = 'Country')

plt.subplots_adjust(wspace=.5)
plt.show()
```
